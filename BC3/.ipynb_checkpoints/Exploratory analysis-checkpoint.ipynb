```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
departments = pd.read_csv("departments.csv")
```

```python
departments
```

```python
departments["department"].unique()
```

```python
orders = pd.read_csv("orders.csv")
```

```python
conditions = [
    (orders['order_dow'] == 0),(orders['order_dow'] == 1),(orders['order_dow'] == 2),(orders['order_dow'] == 3),
    (orders['order_dow'] == 4),(orders['order_dow'] == 5),(orders['order_dow'] == 6),
    ]
```

```python
values = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
```

```python
orders['day_of_week'] = np.select(conditions, values)
```

```python
orders = orders.sort_values("order_dow")
```

```python
orders
```

```python
# quantas vezes encomendaste
len(orders["order_number"].unique())
```

```python
len(orders["order_id"].unique())
```

```python
sns.set_theme(style="whitegrid")
ax = sns.countplot(x="day_of_week", data=orders, palette="Spectral")
ax.set(xlabel='Days of Week', ylabel='Number of Ocurrences', title='Orders per Day of Week')
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
```

```python
sns.set_theme(style="whitegrid")
ax = sns.countplot(x="order_hour_of_day", data=orders, palette="Spectral")
ax.set(xlabel='Hours', ylabel='Number of Ocurrences', title='Orders per Hour of Day')
```

```python
dspo = orders["days_since_prior_order"].dropna().to_frame()
```

```python
dspo = orders["days_since_prior_order"].value_counts().to_frame()
```

```python
dspo
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')

x = dspo.index
y = dspo["days_since_prior_order"]


plt.bar(x, y, color='green')

plt.xlabel("Days Since Prior Order")
plt.ylabel("Number of Ocurrences")
plt.title("Days Since Prior Order")

plt.show()
```

```python
grouped_df = orders.groupby(["day_of_week", "order_hour_of_day", "order_dow"])["order_number"].aggregate("count").reset_index()
grouped_df = grouped_df.sort_values("day_of_week")
grouped_df = grouped_df.pivot("day_of_week", 'order_hour_of_day', 'order_number')

plt.figure(figsize=(12,6))
sns.heatmap(grouped_df, cmap="Spectral")
plt.title("Frequency in Purchases per Hour and Day of Week")

plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")

plt.show()
```
